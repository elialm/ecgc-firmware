```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.fft
```

```python
# Number of samplepoints
N = 100
# sample spacing
T = 1.0 / 50.0

x = np.linspace(0.0, N * T, N)
y = np.sin(2.0 * np.pi * x) + np.sin(2.0 * np.pi * x * 3) + np.sin(2.0 * np.pi * x * 5) + np.sin(2.0 * np.pi * x * 7)

plt.plot(x, y)
plt.show()
```

```python
yf = scipy.fft.fft(y)
xf = np.linspace(0.0, 1.0 / (2.0 * T), N // 2)

plt.plot(xf, 2.0 / N * np.abs(yf[:N // 2]))
plt.show()
```
