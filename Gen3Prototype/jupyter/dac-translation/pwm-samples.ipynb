```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal

plt.rcParams['figure.figsize'] = (20, 10)
```

```python
# Number of samplepoints
N = 4000
# sample spacing
T = 1.0 / 8000000.0

t = np.linspace(0.0, N * T, N)

# Audio wave
fa = np.sin(2.0 * np.pi * t * 2000.0)

# Triangle wave
ftr = 1.1 * signal.sawtooth(200000.0 * 2.0 * np.pi * t, 0.5)

# Modulated PWM
fp = [ 1 if a >= tr else 0 for a, tr in zip(fa, ftr) ]

plt.plot(t, fa, t, ftr, t, fp)
plt.xscale('linear')
plt.xlabel('time (s)')
plt.ylabel('magnitude')
plt.show()
```

```python
OUTPUT_FILE_NAME = 'pwm.csv'

with open(OUTPUT_FILE_NAME, 'w') as out_file:
    for sample in fp:
        out_file.write('{}\n'.format(str(sample * 3.3)))
```
